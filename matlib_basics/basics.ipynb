```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

BASIC GRAPH

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

## Resize graph
plt.figure(figsize=(5,3), dpi=150)

##Shorthand notation for customizing graph
## fmt = '[color][marker][line]

##Line 1
plt.plot(x,y, label='2x', color='red', marker='.', markersize=15)

##Line two
x2= np.arange(0,4,0.25)
plt.plot(x2[:6], x2[:6]**2, label= 'x^2')

##PLot remainder of graph for future predictions
plt.plot(x2[5:], x2[5:]**2, 'b--')


#Styling//Title/X and Y labels
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()

plt.savefig('mygraph.png')

plt.show()
```

Bar Chart

```python
labels = ['A', 'B', 'C']
vals = [1,6,2]

bars = plt.bar(labels,vals)

bars[1].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```
